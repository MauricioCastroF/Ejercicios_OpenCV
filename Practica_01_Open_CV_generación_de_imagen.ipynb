```python
#importar OpenCV
import cv2
#importar Numpy
import numpy as np
#importar matplotlib
import matplotlib.pyplot as plt
```

```python
#crear una imagen negra
img1 = np.zeros((10,10,1), np.uint8)
img2 = np.zeros((20,20,1), np.uint8)
```

```python
# Mostramos nuestra imagen
plt.imshow(img1, cmap='gray')
plt.show()
```

```python
plt.imshow(img2, cmap='gray')
plt.show()
```

```python
#cambiar el color de algunos píxeles
img1[3,2] = 100
img1[4,4] = 32
```

```python
# Mostramos nuestra imagen
plt.imshow(img1, cmap='gray')
plt.show()
```

```python
#cambiar el color de algunos pixeles en la img2
img2[8,1] = 25
img2[8,3] = 200
```

```python
# Mostramos nuestra imagen
plt.imshow(img2, cmap='gray')
plt.show()
```
